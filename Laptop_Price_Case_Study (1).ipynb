```python
def fetch_processor_name(proc):
    if proc.split()[0]=='Intel':  # first we check if the first word in intel
        if proc.split()[1]=='Core':  # secondly we check if the second word in core
            return proc             # then proc name will not be change
        else:
            return " ".join(proc.split()[0:2])  # if second word in not core then we will only take 2 words and join them
    else:
        if proc[4]=="E":    # if fourth word in E
            return "AMD E-Series"
        elif proc[4]=='R':
            return "AMD Ryzen Series"
        elif proc[4]=='A':
            return "AMD A series"
        else:
            return "AMD FX Series"
```

```python
df['Cpu']=df['Cpu'].apply(fetch_processor_name)
```

```python
# Ridge Regression : L2 Regularization : penalty: square of coefficients
ct_ohe= ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])],remainder='passthrough')
reg_model= Ridge(alpha=1)
pipe=Pipeline([('s1',ct_ohe),('s2',reg_model)])

pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print("R2 score:",r2_score(y_test,y_pred))
print("MAE:",mean_absolute_error(y_test,y_pred))
print("RMSE:",root_mean_squared_error(y_test,y_pred))
```

```python
# Lasso Regression  : L1 regularization : penalty : absolute value of coefficients
# It has internal automatic feature selection
# Lasso is more strict
# it string the coefficient and put the unimportant feature to zero
# If you want automatic feature selection to can use Lasso
# it automatically drop useless features by setting the coefficient to zero
# Lasso is useful when you have so many features and you know that all features are not useful
# When you care more about interpretability


# When is Lasso useful?
    # - When you have many features, and you think not all are useful`
    # - you suspect some features are not important
    # - you want a simpler model

# Lasso is very sensitive to feature scaling
# if your features are on very different scales, it is a good idea to normalize or standardize them
# we
```

```python
# Lasso Regression
categorical_features=[0,1,2,4,5]
numerical_features=[3,6,7,8,9,10,11,12]
preprocessor= ColumnTransformer(transformers=[
    ('cat',OneHotEncoder(drop='first'),categorical_features),
    ('num',StandardScaler(),numerical_features)
    ])
reg_model= Lasso(alpha=0.01,max_iter=100000)
pipe=Pipeline([('s1',preprocessor),('s2',reg_model)])

pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print("R2 score:",r2_score(y_test,y_pred))
print("MAE:",mean_absolute_error(y_test,y_pred))
print("RMSE:",root_mean_squared_error(y_test,y_pred))
```

```python
# XG Boost is a smart learning algorithm, which build many decision tree one after another,
# In which one tree work with the reference of other tree's mistakes
# From the mistake of pervious one next particular thing happens
# XGB means Extreme Gradient Boosting, it is called extreme bcoz it is super optimise.
# It imporves using Gradient
# XG boost find the errors made by the first tree.
# Next tree predict the errors of previous one and add the correct prediction
# So over time it builds a better and better prediction by fixing .
# It figure out how much each tree should change to imporve Gradiant Boosting
# It is fast, flexible, optimise quickly , powerful and other things which makes it much much faster
```

```python
# XGB Regression
categorical_features=[0,1,2,4,5]
numerical_features=[3,6,7,8,9,10,11,12]
preprocessor= ColumnTransformer(transformers=[
    ('cat',OneHotEncoder(drop='first'),categorical_features),
    ('num',StandardScaler(),numerical_features)
    ])
reg_model= XGBRegressor(random_state=15,n_estimators=950, learning_rate=0.1,max_depth=2)
pipe=Pipeline([('s1',preprocessor),('s2',reg_model)])

pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print("R2 score:",r2_score(y_test,y_pred))
print("MAE:",mean_absolute_error(y_test,y_pred))
print("RMSE:",root_mean_squared_error(y_test,y_pred))
```

```python
# GradientBoosting Regression
categorical_features=[0,1,2,4,5]
numerical_features=[3,6,7,8,9,10,11,12]
preprocessor= ColumnTransformer(transformers=[
    ('cat',OneHotEncoder(drop='first'),categorical_features),
    ('num',StandardScaler(),numerical_features)
    ])
reg_model= GradientBoostingRegressor(random_state=15,n_estimators=950, learning_rate=0.1,max_depth=2)
pipe=Pipeline([('s1',preprocessor),('s2',reg_model)])

pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print("R2 score:",r2_score(y_test,y_pred))
print("MAE:",mean_absolute_error(y_test,y_pred))
print("RMSE:",root_mean_squared_error(y_test,y_pred))
```

```python
%%writefile laptop_app.py
import streamlit as st
import pickle
df=pickle.load(open('df.pkl','rb'))
pipe=pickle.load(open('pipe.pkl','rb'))
st.title("Laptop Price Predictor App")
st.text("This app is using only a select few laptops(around 1200 laptops), so it may not align exactly with real world data")

company=st.selectbox("Manufacturer of the Laptop",df['Company'].unique(),index=4)  # using index we get pre select in the app or default selection
typename=st.radio("Type of the Laptop",df['TypeName'].unique(),horizontal=True,index=1)
cpu=st.selectbox("Processor",df['Cpu'].unique())
ram=st.pills("RAM on the system(in GB)",[4,8,12,16,24,32,64,128])  # we have make a list of ram by ourself
gpu=st.radio("Graphics Card",df['Gpu'].unique(),index=1,horizontal=True) # making it horizontallyy
os=st.selectbox("Operating System",df['OpSys'].unique(),index=2)
weight=st.slider("Weight of the laptop(in kg)",min_value=0.7,max_value=4.8,step=0.1,value=2.1) # value means default value of slider
touchscreen=st.pills("Does the laptop have touchscreen?",['Yes','No'])
ips=st.pills("Does the laptop have an IPS display?",['Yes','No'])
cpu_speed=st.slider("Clock Speed of CPU(in GHz)",min_value=0.9,max_value=3.6,step=0.1,value=2.3)
hdd=st.pills("Hard disk size on the system(in GB). If only SSD is present, select this is as 0",
             [0,512,1024,2000])
ssd=st.pills("SSD storage on the system(in GB).",[0,256,512,1024,2000])
screen_size=st.slider("Screen size(measured diagonally, in inches)",min_value=10.0,max_value=18.4,value=15.6,step=0.1)
screen_resolution=st.selectbox("Laptop Screen Resolution (in pixels)",
 ["2560x1600","1440x900","1920x1080","2880x1800","1366x768","2304x1440","3200x1800","1920x1200","2256x1504",
  "3840x2160","2160x1440","2560x1440","1600x900","2736x1824","2400x1600"],index=2)

if st.button("PREDICT PRICE"):
    if touchscreen == 'Yes':
        touchscreen=1
    else:
        touchscreen=0
    if ips == 'Yes':
        ips=1
    else:
        ips=0
    X_res=int(screen_resolution.split('x')[0])
    Y_res=int(screen_resolution.split('x')[1])
    ppi = ((X_res**2)+(Y_res**2))**0.5/screen_size
    query=[[company,typename,cpu,ram,gpu,os,weight,touchscreen,ips,cpu_speed,hdd,ssd,ppi]]
    op=pipe.predict(query)
    st.subheader("The estimated price of the laptop with the above mentioned specs is "+str(int(round(op[0],-2)))+".")
```
